```python
# Creating a function to assess the dataframes 
def assess(df):
    print('Checking the shape of {}'.format(df.name))
    print('The dataset has {} rows and {} columns'.format(df.shape[0], df.shape[1]))
    print('+' * 50)
    print('+' * 50)
    print('Checking the number of duplicate values')
    print('{} has {} duplicates'.format(df.name, df.duplicated().sum()))
    print('+' * 50)
    print('+' * 50)
    print('Checking the number of null values')
    print(df.isna().sum())
    print('+' * 50)
    print('+' * 50)
    print('Checking the data types')
    print(df.info())
    print('+' * 50)
    print('+' * 50)   
```

### Correction 2 - remove unnecessary columns

 Since, we are focusing on the departure flights from SFO, we will be removing the arrival columns from runway_df

```python
runway_df = runway_df[runway_df.columns.difference(['01L/R Arrival', '01L/R ArrivalPercentage Utilization', '10L/R Arrival',\
                                           '10L/R ArrivalPercentage Utilization', '19L/R Arrival', \
                                           '19L/R ArrivalPercentage Utilization', '28L/R Arrival', '28L/R ArrivalPercentage Utilization'])]
```

### Correction 3 - Merge the night_df and runway_df datasets together based on the year and month

```python
# merge the datasets
merged_df = pd.merge(night_df, runway_df, on=['Year', 'Month'])
merged_df.head()
```

#### Rename the columns


```python
# rename the column names to a more conscise and understandable one
merged_df = merged_df.rename(columns = {'01L/R': 'night_01L/R', '01L/R Percent of Departures': 'night_01L/R_percent', '10L/R': 'night_10L/R',\
                  '10L/R Percent of Departures': 'night_10L/R_percent', '19L/R': 'night_19L/R', '19L/R Percent of Departures': 'night_19L/R_percent',\
                  '28L/R': 'night_28L/R', '28L/R Percent of Departures': 'night_28L/R_percent', '01L/R Departure': 'total_01L/R',\
                  '01L/R DeparturePercentage Utilization': 'total_01L/R_percent', '10L/R Departure' : 'total_10L/R',\
                  '10L/R DeparturePercentage Utilization' : 'total_10L/R_percent', '19L/R Departure': 'total_19L/R',\
                  '19L/R DeparturePercentage Utilization': 'total_19L/R_percent', '28L/R Departure': 'total_28L/R',\
                  '28L/R DeparturePercentage Utilization': 'total_28L/R_percent'})
```

```python
merged_df.columns = merged_df.columns.str.lower()
```

```python
runway_night = ['night_01l/r', 'night_10l/r', 'night_19l/r', 'night_28l/r']
runway_total = ['total_01l/r', 'total_10l/r', 'total_19l/r', 'total_28l/r']
```

## Total number of night flights per year in SFO from 2005 - 2019

```python
# Group the data by year and sum the values of runway_night columns for each year
night_flights_by_year = merged_df[runway_night + ['year']].groupby('year').sum()

# Set the size of the plot
plt.figure(figsize=(12, 8))

# Set the line color and style
plt.plot(night_flights_by_year.index, night_flights_by_year.values, linewidth=2, label = ['01L/R', '10L/R', '19L/R', '28L/R'])

# Add a title and axis labels
plt.title('Night Flights per Year', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Night Flights', fontsize=14)

# Customize the tick labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Set the background color
plt.gca().set_facecolor('white')

# set legend
plt.legend(loc='upper left', fontsize=12)

# Display the plot
plt.show()
```

## Top 10 airlines with the highest average "Noise Exceedance Rating"
These airlines produce lesser noise pollution.

NOte: There are inconsistency with the dataset. Some airlines are represented by more than one name e.g. NETJETS is represented by both NETJETS and NETJETS AVIATION. Hence, the airline code will be use instead for the analysis.

```python
top_airlines = airline_df[['Noise Exceedance Rating', 'Airline Code']].groupby('Airline Code').mean().sort_values(by='Noise Exceedance Rating', ascending=False)[:10]
top_airlines.plot(kind='bar', figsize=(10, 8))
plt.ylim(9.8, 10) # Set the y-axis limits to 9.8 and 10
plt.title('Top 10 Airlines that produces least noise pollution')
plt.xticks(rotation = 0)
plt.xlabel('Airline')
plt.ylabel('Mean Noise Exceedance Rating')
plt.show()
```

The above 10 airlines are recommended for a more effective noise pollution management.

## Airlines having their average "Noise Exceedance Rating" less than 5.0
These airlines produce greatest noise pollution.

```python
top_airlines = airline_df[['Noise Exceedance Rating', 'Airline Code']].groupby('Airline Code').mean().sort_values(by='Noise Exceedance Rating', ascending=True)
top_airlines = top_airlines[top_airlines['Noise Exceedance Rating'] < 5]
top_airlines.plot(kind='bar', figsize=(12, 8))
# plt.ylim(8, 10) # Set the y-axis limits to 8 and 10
plt.title('Top 10 Airlines that produces greatest noise pollution')
plt.xticks(rotation = 0)
plt.xlabel('Airline')
plt.ylabel('Mean Noise Exceedance Rating')
plt.show()
```

> The above airlines produce the greatest noise pollution. Bans could probably be placed on the airlines until they are able to provide better aircrafts with noise-reduction control system.
